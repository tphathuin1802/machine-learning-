```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo một chuỗi thời gian với 100 điểm dữ liệu
time = np.linspace(0, 10, 100)

# Mô phỏng một cyclical pattern (biến động chu kỳ dài hạn)
# Sử dụng hàm sin để tạo sự biến động với chu kỳ không đều và thay đổi theo thời gian
cyclical_pattern = 5 * np.sin(2 * np.pi * time / 5) + np.random.normal(scale=0.5, size=len(time))

# Vẽ đồ thị
plt.figure(figsize=(10, 6))
plt.plot(time, cyclical_pattern, label="Cyclic Pattern", color='b')

# Thêm tiêu đề và nhãn
plt.title('Cyclic Pattern Example in Time Series', fontsize=16)
plt.xlabel('Time', fontsize=14)
plt.ylabel('Value', fontsize=14)

# Hiển thị đồ thị
plt.legend()
plt.grid(True)
plt.show()
```
