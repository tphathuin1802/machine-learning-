```python
import numpy as np
import matplotlib.pyplot as plt

dataset = np.array([
    [1, 1, 0], [2, 1, 0], [3, 1, 0], [4, 1, 0],
    [1, 2, 0], [2, 2, 0], [3, 2, 0], [4, 2, 0],
    [1, 3, 1], [2, 3, 1], [3, 3, 0], [4, 3, 0],
    [1, 4, 1], [2, 4, 1], [3, 4, 0], [4, 4, 0],
    [1, 5, 0], [2, 5, 0], [3, 5, 0], [4, 5, 0],
    [1, 6, 0], [2, 6, 0], [3, 6, 0], [4, 6, 0],
])
print(dataset)
```

```python
group_a = dataset[dataset[:,2]==0]
group_b = dataset[dataset[:,2]==1]
plt.scatter(group_a[:,0 ], group_a[:,1], color = "blue", label = "Group A")
plt.scatter(group_b[:,0], group_b[:,1], color = "red", label = "Group B")
plt.title("Mouse Dataset")
plt.xlabel("Front leg's  length")
plt.ylabel("Back leg's length")
plt.legend(loc="upper left")
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
x = dataset[:, :2]
y = dataset[:, 2]

decision_tree = DecisionTreeClassifier(criterion="gini")
decision_tree.fit(x,y)


```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,2)
plot_tree(decision_tree, fontsize= 12)
plt.subplot(1,2,1)
plt.scatter(group_a[:, 0], group_a[:, 1], color="blue", s=100, label='Group A')
plt.scatter(group_b[:, 0], group_b[:, 1], color="red", s=100, label="Group B")
plt.title("Mouse Classification")
plt.xlabel("Front Leg's length")
plt.ylabel("Back Leg's length")
plt.legend()
plt.show()
```
