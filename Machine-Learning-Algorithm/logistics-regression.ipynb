## Logistic Regression


Function:
 y= 1/ b ^ -(z)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create a dataset
dataset = np.array([
    [-10, 0],
    [-5, 0],
    [-7, 0],
    [0, 0],
    [-2, 0],
    [5, 1],
    [7, 1],
    [6, 1],
    [10, 1],
    [15, 1],
    [9, 1]
])
```

```python
negative_0 = dataset[:5]
positive_1 = dataset[5:]

plt.scatter(negative_0[:,0], negative_0[:,1], c="y")
plt.scatter(positive_1[:,0], positive_1[:,1], c="g")
plt.show()
```

```python
import numpy as np

def get_prediction(m, b, x):
    """Computes the logistic function (sigmoid) for predictions"""
    return 1 / (1 + np.exp(-(m * x + b)))

def get_cost(y, y_hat):
    """Calculates the cost function (binary cross-entropy loss)"""
    k = y.shape[0]
    return (-1/k) * np.sum(y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat))

def get_gradient(m, b, x, y, y_hat):
    """Calculates the gradients of the cost function w.r.t m and b"""
    k = y.shape[0]
    dm = (1/k) * np.sum((y_hat - y) * x)
    db = (1/k) * np.sum(y_hat - y)
    return dm, db

def get_accuracy(y, y_hat):
    """Calculates the accuracy of the model"""
    return ((y_hat >= 0.5).astype(int) == y).sum() / y.shape[0]

# Parameters
m = 1.0  # Initial weight (slope)
b = 0.0  # Initial bias
iterations = 150
lr = 0.03  # Learning rate

# Example dataset (replace this with your actual dataset)
# For example, a simple linearly separable dataset
# x = feature, y = true label
# Create a dataset
dataset = np.array([
    [-10, 0],
    [-5, 0],
    [-7, 0],
    [0, 0],
    [-2, 0],
    [5, 1],
    [7, 1],
    [6, 1],
    [10, 1],
    [15, 1],
    [9, 1]
])
  # Example dataset
x = dataset[:, 0]
y = dataset[:, 1]

# Gradient descent loop
costs = []

for iter in range(iterations):
    # Get the predictions (y_hat) for current parameters m and b
    y_hat = get_prediction(m, b, x)

    # Compute the cost using current predictions
    cost = get_cost(y, y_hat)

    # Compute the gradients
    dm, db = get_gradient(m, b, x, y, y_hat)

    # Update parameters using gradient descent
    m -= lr * dm
    b -= lr * db

    # Calculate accuracy
    accuracy = get_accuracy(y, y_hat)

    # Print iteration details
    print(f"Iteration {iter + 1} - Cost: {cost}, Accuracy: {accuracy * 100}%")

	#update parameter

    # Optionally store the cost for plotting
    costs.append(cost)
```
